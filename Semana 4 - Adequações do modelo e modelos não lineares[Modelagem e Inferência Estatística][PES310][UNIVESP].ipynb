**ADEQUA√á√ÉO DO MODELO: EXEMPLO 1**

Dados sobre x = taxa de libera√ß√£o na √°rea do queimador e y = emiss√µes de NOx. Observe os dados e os valores ajustados, os res√≠duos e os res√≠duos
padronizados. A reta de regress√£o estimada √© $y = ‚Äì45,55 + 1,71x$ e $r^2 = 0,961$. 

```python
lstx2 = (100, 125, 125, 150, 150, 200, 200, 250, 250, 300, 300, 350, 400, 400)
lsty2 = (150, 140, 180, 210, 190, 320, 280, 400, 430, 440, 390, 600, 610, 670)
# Construir o DataFrame e nomear as colunas
df = pd.DataFrame(list(zip(lstx2, lsty2)),
                  columns =["x","y"])
x=df ['x']
y=df ['y']
```

image.png

```python
#adicionar uma constante preditora
x = sm.add_constant(x)
#fit linear regression model
modelo = sm.OLS(y, x).fit() #aten√ß√£o esta vari√°vel ser√° usada no IC
print(modelo.summary())
```

```python
# par√¢metros da reta
b=modelo.params
b0=b[0] #intercepto
b1=b[1] #inclina√ß√£o
print('y= {0}{1} x'.format(b0,b1))
```

```python
y_pred=list(modelo.predict())
#sns.resid(modelo)
resi=list(modelo.resid)
```

```python
#crear instancia influence
influence = modelo.get_influence()
#obter residuos standardizados
stdresid = list(influence.resid_studentized_internal)
```

```python
prop=np.divide(resi,stdresid) #e/e* para verificar a proporcionalidade entre o erro e o erro padronizado
```

```python
#@title Tabela geral
dft = pd.DataFrame(list(zip(lstx2, lsty2, y_pred, resi, stdresid, prop)),
               columns =["x","y","yc","e","e*","e/e*"])
dft 
```

e* n√£o √© m√∫ltiplo de e devido ao fato de que as vari√¢ncias residuais s√£o diferentes umas das outras.
