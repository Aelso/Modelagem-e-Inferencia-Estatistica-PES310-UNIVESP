**TESTE HIPÓTESE**

Teste de utilidade do modelo, isto é, confirmar ou rejeitar sua utilidade.
Hipóteses nula  $H_0:\beta_1= \beta_10$

Estatística de teste: $t=\frac{\hat{\beta}_1-\beta_{10}}{s_{\hat{\beta}_1}}$ 

 |Hipóteses alternativa    |Determinação do valor -p                       |
 |-------------------------|-----------------------------------------------|
 |$H_a: \beta_1 >\beta_10$ |Área sob a curva $t_{n-2}$ á direita de t.     |
 |$H_a: \beta_1 <\beta_10$ |Área sob a curva $t_{n-2}$ á esquerda de t.    |
 |$H_a: \beta_1 ≠\beta_10$ |2.(Área sob a curva $t_{n-2}$ á direita de|t|) |


O teste de utilidade do modelo é o teste de  $H_0: \beta_1 = 0$ versus $H_a: \beta_1 ≠\beta_0$, sendo o valor da estatística de teste a razão $t={\hat{\beta}_1/ S_{\hat{\beta}_0}}$

 Fonte: (DEVORE, 2018, p. 482)

**PROCEDIMENTOS PARA O TESTE DE HIPÓTESE**

1. Obter a reta de regressão $y=\beta_0 + \beta_1 x$
2. Definir o valor de n (número de amostras).
3. Definir o valor de k (número de variáveis).
4. Calcular os graus de liberdade gl = n-k.
5. Se a intenção é testar a utilidade do modelo iniciar testando $H_0:\beta_1= 0$ frente a  $H_a: \beta_1 ≠ 0$ depois testar  $H_0: \beta_0 = 0$ frente a  $H_a: \beta_0 ≠ 0$ 

**PROCEDIMENTOS PARA O TESTE DE HIPÓTESE $\beta_1$**

1. Definir a hipótese nula $H_0:\beta_0= 0$  frente a $H_a: \beta_10 ≠ 0$.
2. Definir o intervalo de confiança α se conhecido.
3. Definir o intervalo crítico $t_{crit}$ na tabela t-student para um determinado α e gl.
4. Definir a estatística de teste $t=\frac{\hat{\beta}_1-\beta_{10}}{s_{\hat{\beta}_1}}$ 
como  $H_a: \beta_10 ≠ 0$ ->$t=\frac{\hat{\beta}_1}{s_{\hat{\beta}_1}}$ 
5. Se |t|≥$t_{crit}$  rejeitar $H_0$ em favor de $H_a: \beta_10 ≠ 0$
6. Caso contrario comparar se p-valor < α.


**PROCEDIMENTOS PARA O TESTE DE HIPÓTESE $\beta_0$**

1. Definir a hipótese nula $H_0:\beta_0= 0$  frente a $H_a: \beta_0 ≠ 0$.
2. Dos dados do problema $1- α= 0,95 -> α=0,05$ e $α/2=0,025$ 
3. Definir o intervalo crítico( $t_{crit}$) na tabela t-student para um determinado α e gl.
4.Definir a estatística de teste $t=\frac{\hat{\beta_0}-\beta_{00}}{s\sqrt(\frac{1}{n}+\frac{\bar{x}^2}{S_{XX}})}$ .

  Como  $H_0: \beta_0 = 0$ -> $t={\hat{\beta_0}} /\sqrt(\frac{1}{n}+\frac{\bar{x}^2}{S_{XX}})$ 

5. Se |t|≥$t_{crit}$  rejeitar $H_0$ em favor de $H_a: \beta_0 ≠ 0$
6. Caso contrario comparar se $2(p-valor) < α$ se sim, rejeitar $H_0$ em favor de $H_a: \beta_0 ≠ 0$.

**INFERÊNCIAS SOBRE $\mu_{y, x*}$**

Seja $\hat{Y} = \hat{\beta}_0-\beta_{1 X^ *}$ , em que x* é algum valor fixo de x. Então 
1. O valor médio de $\hat{Y}$ é 

$E(\hat{Y}) = \hat{\beta}_0-\beta_{1 X^ *}= \mu_ {\hat{\beta}_0-\beta_{1 X^ *}} = \hat{\beta}_0-\beta_{1 X^ *}$
Assim, $\hat{\beta}_0-\beta_{1 X^ *}$ 

é um estimador não viciado para $\hat{\beta}_0-\beta_{1 X^ *}$ (isto é, para $\mu_ {\hat{\beta}_0}-\beta_{1 X^ *}$).




2. A variância de $\hat{Y}$ é 

image.png

e o desvio padrão    $ α _ \hat{Y}$ é a raiz quadrada dessa expressão.
O desvio padrão estimado de $ \hat{\beta}_0-\beta_{1 X^ *}$, representado por $ S_{\hat{\beta}_0-\beta_{1 X^ *}}$ resulta da substituição de α por sua estimativa s:

$S_{\hat{Y}}= s_{\hat{\beta}_0-\hat{\beta_{1 X^ *}}}= S\sqrt{(\frac{1}{n}+\frac{({x^ *- \hat{x}}^2)}{S_{xx}})}$ 

3. $\hat{Y}$ tem uma distribuição normal.

Definir uma variável T para obter a estatística de teste para reta

$T= \frac{\hat{\beta}_0-\hat{\beta}_{1 X^ *} -{\beta}_0+\beta_{1 X^ *}} {S_{\hat{\beta}_0+\hat{\beta}_{1 X^ *}}}$


Considerando gl=n-2

**INTERVALO DE PREVISÃO PARA UM VALOR FUTURO DE Y**

Intervalo de confiança (IC).
> Referido a um parâmetro ou população com valor fixo.

>  Valor desconhecido para o pesquisador.

Intervalo de previsão (IP).
> Valor futuro de Y.

> Variável aleatória.

$V[Y- \hat{\beta}_0+\hat{\beta}_1X^ *]=$ variância do erro de previsão 

$V(Y)+V(\hat{\beta}_0+\hat{\beta}_1X^ *)$

$\alpha ^2 + \alpha^2[{\frac{1} n} + \frac{({x^ *} - ̂{\bar{x}})^ 2} {S_xx}]$

$ \alpha^2[1+{\frac{1} n} + \frac{({x^ *} - ̂{\bar{x}})^ 2} {S_xx}]$

$ T= \frac{Y-\hat{\beta}_0+\hat{\beta}_{1X^ *}}{2 \sqrt{{1+\frac{1} n} + \frac{({x^ *} - ̂{\bar{x}})^ 2} {S_xx}}}$

Um IP de $100(1 – a)\%$ para uma observação Y futura a ser feita quando $x = x^*$ é

$\hat{\beta}_0-\hat{\beta}_{1 X^ *} ± t_\frac{α} 2$,$n-2*{S \sqrt{{1+\frac{1} n} + \frac{({x^ *} - ̂{\bar{x}})^ 2} {S_xx}}}$

$\hat{\beta}_0-\hat{\beta}_{1 X^ *} ± t_\frac{α} 2$,$n-2* \sqrt{{S^ 2} + S^ 2_{\hat{\beta}_0-\hat{\beta}_{1 X^ *}}}$







$\hat{y} ± t_{\frac{α} 2,n-2}*\sqrt{{S^ 2} + S^ 2_\hat{Y}}$

Fonte: (DEVORE, 2018, p. 491-492)
