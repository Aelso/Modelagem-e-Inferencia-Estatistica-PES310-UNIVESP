normal $ùúá_ùúÄ = 0$ e $ùúé_ùúÄ^2 = 0$


$b_0n + b_1 Œ£x_i$

$b_0n + b_1 Œ£x_i + b_2 Œ£x_i^2 +...+ b_k Œ£x_i^k = Œ£y_i$

$b_0Œ£x_i + b_1 Œ£x_i^2 + b_2 Œ£x_i^3+...  + b_k Œ£x_i^{k+1}= Œ£x_iy_i$

**VARI√ÇNCIA ESTIMADA E COEFICIENTE DE DETERMINA√á√ÉO**

Para modelos lineares e a estimativa de $\alpha^2$ √©: $\alpha^2 = s^2 = \frac{SQE}{n-2}$

Para os modelos polinomiais: $\alpha^2 = s^2 = \frac{SQE}{n-(k+1)}$

Coeficiente de determina√ß√£o m√∫ltipla: $R^2 = 1 - \frac{SQE}{SQT}$ 

Coeficiente de determina√ß√£o m√∫ltipla ajustado:

$R^2(ajustado) = 1 - \frac{n-1}{n-(k+1)} . \frac{SQE}{SQT} =   \frac{(n-1)R^2 -k}{(n-1-k}$

b) Que propor√ß√£o da varia√ß√£o observada na densidade pode ser
atribu√≠da √† rela√ß√£o do modelo?

$R^2 = 1 - \frac{SQE}{SQT}$  
 
$R^2(ajustado) = 1 - \frac{n-1}{n-(k+1)} . \frac{SQE}{SQT} =   \frac{(n-1)R^2 -k}{(n-1-k}$ 

```python
#@title resposta b)
#calcular SQE--> Baseado nos valores esperados
sqe = np.sum((ypred - y)**2)
print("SQE=", sqe)

#calcular SQT-->Baseado nos valores observados
sqt = np.sum((y - y.mean())**2)
print("SQT=", sqt)

#calcular SQR
sqr = sqt - sqe
print("SQR=",sqr)

R2=1-sqe/sqt
print("Coeficiente de determina√ß√£o m√∫ltippla=",R2, "ou",R2*100, "%")

k=modelpo.df_model # grau do modelo
n=modelpo.nobs # num. amostras
R2adj=((n-1)*R2-k)/(n-1-k)

print("R2 ajustado=",R2adj, "ou",R2adj*100, "%")
```
