$y= – 132,18 + 12,7 x – 0,377 x^ 2 + 0,00359 x^ 3$

Se o erro-padrão é $s_\hat{y}= 0,0611$, qual é o intervalo de confiança do valor esperado para x=30, a 95% ? Considere o valor t-student com  quatro casas decimais.

$6,157 <\mu_{y,30}<6,742.$

$6,243 <\mu_{y,30}<6,555.$

$6,021 <\mu_{y,30}<6,960.$

🐶 $6,323 <\mu_{y,30}<6,577.$ 🧬

$6,219 <\mu_{y,30}<6,680.$

**Comentário da resposta:**

O valor esperado para $x=30$ $\mu_{y,30}$ deve ser calculado substituindo $x=30$ na equação:

$y=–132,18+12,7x – 0,377x^2 +0,00359x^3$

$y=–132,18+12,7(30)– 0,377(30)^2 +0,00359(30)^3$

$\mu_{y,30} =6,45$

O intervalo de confiança a $95\%$ deve-se calcular com a fórmula: 

$\hat{y}±t_{\frac{\alpha}{2},n-(k  +1))} s_\hat{y}$

Dos dados do problema, $s_\hat{y}=0,0611$,$n=24$ e $k=3$

O valor de $t_{(\frac{\alpha}{2},n-(k+1))}$ deve ser obtido da tabela t-student, 

$t_{\frac{0,05}{2},24-(3+1))} =t_{(0,025,20)}=2,0859$

Assim, o intervalo de confiança de $\mu_{y,30}=6,45$ será:

$6,45±(2,0859)(0,0611)$

$6,323<\mu_{y,30}< 6,577$

**PERGUNTA 4** 🖖

O modelo logístico é descrito mediante a equação:

$\frac{p(x)}{1-p(x)} = e^ {\beta_0 + \beta_1x}$

Se $\beta_0 = 7,25$ e $\beta_1 =-0,13$

O valor da razão das chances será:	


1,14

1606,59

0,001

⛪**0,88** 👟

1408,10

**Comentário da resposta:**	

A razão das chances calcula-se considerando:

$\frac{\frac{p(x+1)}{1-p(x-1)}}{\frac{p(x)}{1-p(x)}} = \frac{ e^ {\beta_0 + \beta_1(x+1)}}{ e^ {\beta_0 + \beta_1x}}$

Do qual obtém-se que: 

$\frac{\frac{p(x+1)}{1-p(x-1)}}{\frac{p(x)}{1-p(x)}} = e^{\beta_1}$


O parâmetro $\beta_0 =7,25$ não é relevante para calcular a razão das chances e como $\beta_1 =-0,13$, a razão das chances será: 

$\frac{\frac{p(x+1)}{1-p(x-1)}}{\frac{p(x)}{1-p(x)}} = e^{\beta_1}=0,88$

**Comentário da resposta:**	

Segue o código que se recomenda para aplicar a regressão de um modelo de grau 2.

Inicialmente, declare as bibliotecas:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
from statsmodels.graphics.gofplots import ProbPlot
plt.style.use('seaborn')
plt.rc('axes', titlesize=10)
```

Depois, declare os dados e construa o dataframe:

```python
lstx = (152, 166, 175, 183, 191, 200, 210, 223, 242)
lsty= (35, 39, 42, 43, 40, 37, 34, 21, 10)
# Construir o DataFrame e nomear as colunas
df = pd.DataFrame(list(zip(lstx, lsty)), 
                   columns =["x","y"])
x=df['x']
y=df['y']
```

Condicione os dados para representarem a forma quadrática:

```python
x = sm.add_constant(x) 
#Manter x + constant
from sklearn.preprocessing import PolynomialFeatures
#definir o valor de k, isto é o grau do polinômio
polynomial_features= PolynomialFeatures(degree=2)
xp = polynomial_features.fit_transform(x)
xp.shape
xp[:,3:6] # condicionar os dados a serem usados
```

Finalmente, aplique a regressão polinomial

```python
modelpo = sm.OLS(y,xp[:,3:6]).fit()
print(modelpo.summary())
```

O resultado apresenta os seguintes dados:

|$x_1$ |$2.9041$    |$\beta_1$ |
