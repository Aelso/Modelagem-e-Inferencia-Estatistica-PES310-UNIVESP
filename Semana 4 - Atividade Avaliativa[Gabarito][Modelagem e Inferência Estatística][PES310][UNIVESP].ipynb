```python
#bibliotecas 
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
#Dados
lstx = (1326,1593,4414,5673,29516,26,843,1016,3410,
7101,7356,7904,79,4175,34676,114789,2672,7532,30220)
lsty= (1.495,1.470,1.100,1.190,0.873,1.819,0.810,0.801,
0.600,0.575,0.576,0.580,1.212,0.782,0.596,0.600,0.880,0.883,0.676)
vtx=np.log(lstx)
# Construir o DataFrame e nomear as colunas
df = pd.DataFrame(list(zip(vtx, lsty)),
                  columns =["ln_x","y"])
vtx=df['ln_x']
y=df['y']
#Regress√£o linear com as vari√°veis transformadas
vtx  = sm.add_constant(vtx)
modelo = sm.OLS(y, vtx ).fit() #aten√ß√£o esta vari√°vel ser√° usada no IC
# par√¢metros da reta
b=modelo.params
b0=b[0] #intercepto
b1=b[1] #inclina√ß√£o
print('Fun√ß√£o intrinsecamente linear y= {0}{1} log(x)'.format(b0,b1))
print('Fun√ß√£o linearizada y= {0}{1} x`'.format(b0,b1))
```
